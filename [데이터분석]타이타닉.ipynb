## 1.EDA

### 1-1.결측현항

```python
test_df.info()
```

> Age와 Cabin은 다수의 결측치 처리가 필요하다.  
Fare는 1개 누락임으로 데이터상황에따라 처리여부를 판단

PassengerId : 각 승객의 고유 번호  
Survived : 생존 여부(종속 변수)  
0 = 사망  
1 = 생존  
  
Pclass : 객실 등급 - 승객의 사회적, 경제적 지위  
1st = Upper  
2nd = Middle  
3rd = Lower  
  
Name : 이름  
Sex : 성별  
Age : 나이  
SibSp : 동반한 Sibling(형제자매)와 Spouse(배우자)의 수  
Parch : 동반한 Parent(부모) Child(자식)의 수  
Ticket : 티켓의 고유넘버  
Fare : 티켓의 요금  
Cabin : 객실 번호  
  
Embarked : 승선한 항  
C = Cherbourg  
Q = Queenstown  
S = Southampton  

### 1-2.Label상세

```python
test_df['PassengerId'].value_counts().reset_index()
```

> PassengerId 는 유니크한 값들이다

```python
test_df['Pclass'].value_counts().reset_index()
```

> Pclass 중 가장 3등급이 가장 저렴할것으로 추측된다. 중간인 mid가 오히려 가장 낮다

```python
test_df['Name'].value_counts().reset_index()
```

> Name 는 유니크한 값들이다

```python
test_df['Sex'].value_counts().reset_index()
```

> 성별은 2개 이며 male의 비중이 더 높다

```python
test_df['Age'].value_counts().reset_index()
```

```python
test_df['Age_Level']=test_df['Age']//10*10 # 연령대를 구하기위해 신규 label 생성
```

```python
test_df['Age_Level'].value_counts().reset_index()
```

> 20대가 가장많고, 30대 40대순으로 탑승객이 많다. 10살 이하의 어린이도 발견된다

```python
test_df['SibSp'].value_counts().reset_index() # 형제자매 .배우자
```

```python
test_df['Parch'].value_counts().reset_index() # 부모 자녀
```

> 동반자없이 혼자 탑승한 손님의 비중이 높다 

```python
test_df['Ticket'].value_counts().reset_index() 
```

> Ticket 번호가 중복인것들이 발견된다. 어떤 의미인지 해석 필요

```python
test_df['Fare'].value_counts().reset_index() 
```

> Pclass 와  Embarked(탑승지) 에 따라 가격이 어떻게 나타나는지 확인 필요

```python
test_df['Cabin'].value_counts().reset_index() 
```

> 객실번호가 고객수보다 적다 = 한방에 탑승한 고객들이 존재한다. 동반자일지, 아닐지도 확인필요

```python
test_df['Embarked'].value_counts().reset_index() 
```

> Southampton 탑승자가 가장 많다. 탑승지에 따라 Pclass, 성별, 나이, 요금등 상세한 분석이 가능하다
